### **Git ve GitHub**

### **1. Git ve GitHub Nedir?**

**Git**, yazılım geliştirme süreçlerinde versiyon kontrolünü sağlayan dağıtık bir sistemdir. Her geliştirici, projenin tam geçmişini yerelinde saklar ve paralel çalışma imkânı bulur.

**GitHub** ise Git depolarını barındıran, işbirliği odaklı bir bulut platformudur. Kod paylaşımı, sorun takibi ve otomatik testler gibi özellikler sunar.

| Özellik        | Git                      | GitHub                       |
|----------------|--------------------------|------------------------------|
| Tür            | Versiyon Kontrol Sistemi | Kod Barındırma Platformu     |
| Erişim         | Yerel/Bulut              | Bulut Tabanlı                |
| Kullanım Alanı | Kod Geçmişi Yönetimi     | Ekip İşbirliği ve Paylaşım   |

**GitHub Temel Kavramlar**

| Kavram       | Açıklama                                                                  | Rolü                                                                 |
|--------------|---------------------------------------------------------------------------|----------------------------------------------------------------------|
| Repository   | Proje dosyalarının ve geçmiş değişikliklerinin saklandığı depo             | Projenin merkezi yönetimi ve sürüm kontrolü                           |
| Branch       | Depo içinde bağımsız geliştirme hattı                                     | Farklı özelliklerin ana koda dokunmadan geliştirilmesi                 |
| Commit       | Depodaki belirli bir andaki değişikliklerin kaydı; kısa açıklama içerir | Projenin evrimini kayıt altına alma ve gerektiğinde geri dönme         |
| Push         | Yerelde commit’lenen değişikliklerin GitHub’a gönderilmesi                | Ortak çalışma ve güncel kalma                                        |
| Pull         | Uzak depodaki değişikliklerin yerelden indirilmesi                       | Senkronizasyon ve çakışma yönetimi                                   |

### **2. Temel Avantajlar**

**Değişiklik İzleme:** Her commit (kayıt), kimin neyi ne zaman değiştirdiğini kaydeder.

**Paralel Geliştirme:** Dallar (branch) ile aynı anda birden fazla özellik üzerinde çalışılabilir.

**Geri Alma:** Yanlış değişiklikler kolayca geri alınabilir.

**İşbirliği:** GitHub üzerinden kod incelemeleri (pull request) ve sorun takibi (issue) yapılabilir.

**GitHub Hesabı Oluşturma**

- Web tarayıcınızda https://github.com adresine gidin.

- Ana sayfadaki “Sign up for GitHub” butonuna tıklayın.

- E‑posta, kullanıcı adı ve güçlü bir şifre belirleyip ilerleyin.

- E‑posta doğrulamasını tamamlayın.

- Profil (isim, biyografi vb.) bilgilerinizi isteğe bağlı olarak doldurun.

### **3. GitHub Web Arayüzü ile Başlangıç**

**Depo (Repository) Oluşturma**

- GitHub'da sağ üstteki + > New Repository'ye tıklayın.

-  Depo adını yazın; kısa ve anlamlı bir isim seçin.

- Açıklama ekleyin (isteğe bağlı). Depoyu kimse görmese de açıklama projenizi hatırlatır.

- README oluşturma, .gitignore veya lisans ekleme gibi seçenekleri yapılandırın. İlk defa depo oluştururken README eklemek genellikle önerilir.

- Create repository butonuna tıklayarak işlemi tamamlayın.

#### **3.2. Dosya Yönetimi**

**Dosya Ekleme:** Add file > Upload files ile dosya yükleyin veya doğrudan düzenleyin.



**Commit:** Değişiklikleri kaydetmek için açıklama yazıp Commit changes'e basın.

#### **3.3. Dallarla Çalışma**

Projelerde paralel geliştirme için branch (dal) kullanımı çok önemlidir. Genellikle ana (main veya master) dal değişmeden yeni özellikler veya düzeltmeler yan dallarda yapılır.

- Branch dropdown'ından yeni dal adı girip Create branch'e tıklayın.

- Yeni dalda değişiklik yapıp commit edin.

- Pull Request açarak ana dala birleştirme isteği gönderin.





### **4. GitHub Desktop ile Profesyonel Yönetim**

#### **4.1. Kurulum ve Ayarlar**

##### **İndirme:**

GitHub Desktop resmi sitesinden işletim sisteminize uygun sürümü indirin.

##### **Kurulum:**

İndirilen dosyayı çalıştırıp talimatları takip edin.

##### **Oturum Açma:**

* Uygulamayı açın → **Sign in to GitHub.com**.
* Tarayıcıdan kimlik doğrulamasını tamamlayın.

##### **Git Kimliği Tanımlama:**

* `File` → `Options` (Windows) / `Preferences` (macOS) → `Git`
* `Name` ve `Email` alanlarını doldurun (Git commit'lerinizde görünecek).


#### **4.2. Depo Klonlama ve Senkronizasyon**

##### **Klonlama:**

##### **Web Arayüzü Üzerinden:**

* GitHub'da istediğiniz repo sayfasına gidin → **Code** ▼ → **Open with GitHub Desktop**.
* Uygulama otomatik olarak açılır → Yerel klasör seçip **Clone**'a tıklayın.

##### **Desktop Üzerinden Manuel Klonlama:**

* `File` → `Clone Repository` → `URL`
* Repo linkini yapıştırın (Örn: `https://github.com/kullanici/repo.git`) → **Clone**.

##### **Değişiklikleri Gönderme (Push):**

- Dosyalarda değişiklik yapın.
- Sol paneldeki **Changes** sekmesinde ilgili dosyaları işaretleyin.
- Sağ alttaki kutucuğa **anlamlı bir commit mesajı** yazın.
- **Commit to main** (veya aktif branch adı) → **Push origin** ile GitHub'a gönderin.

##### **Güncellemeleri Çekme (Pull):**

Pull Request (çekme isteği), bir dalda yaptığınız değişiklikleri ana dal veya başka bir dala birleştirmeden önce proje sahiplerinin incelemesine sunar.

`Fetch origin` ile uzak depodaki değişiklikleri kontrol edin.

`Pull origin` ile yerel deponuzu güncelleyin.

#### **4.3. Dallar ve Birleştirme**

##### **Yeni Dal(Brunch) Oluşturma:**

- Üst menüden **Branch** → **New Branch** seçin.
- Dal adı girin (Örn: `feature/login`).
- Base branch'i seçin (Genelde `main`).
- **Create Branch**'e tıklayın.

##### **Pull Request (PR) Açma:**

- Yeni dalda değişiklikleri commit edip **Push origin** yapın.
- GitHub Desktop'ta **Branch** → **Create Pull Request**'e tıklayın.
- Açılan tarayıcıda:
    * **Base branch** (hedef) ve **Compare branch** (kaynak) kontrol edin.
    * Başlık ve açıklama ekleyip **Create pull request**'e basın.
